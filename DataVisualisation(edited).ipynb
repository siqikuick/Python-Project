```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Confidence'], kde=True, color='skyblue')
sns.kdeplot(df['Confidence'], color='red', linestyle='--', label='Kernel Density Estimate')

confidence_interval = np.percentile(df['Confidence'], [2.5, 97.5])
lower_bound, upper_bound = confidence_interval

# Add confidence interval lines to the plot
plt.axvline(lower_bound, color='green', linestyle=':', label='95% Confidence Interval')
plt.axvline(upper_bound, color='green', linestyle=':')

plt.xlabel('Confidence Values')
plt.ylabel('Number of Sentiments')
plt.title('Confidence Interval Plot')

plt.legend()

plt.show()
```

```python
# Define a function to modify sentiment values based on confidence
def modify_sentiment(row):
    sentiment = row['Sentiment']
    confidence = row['Confidence']
    if confidence < 0.75:
        return 'Neutral'
    elif confidence >= 0.75 and sentiment==str('Positive'):
        return 'Positive'
    else:
        return 'Negative'

# Create a new 'Modified Sentiment' column without overwriting 'Sentiment'
df['Modified Sentiment'] = df.apply(modify_sentiment, axis=1)

# Create a DataFrame 'modified_df' containing 'Modified Sentiment' and 'Confidence'
modified_df = df[['Modified Sentiment', 'Confidence']]

# Now, 'modified_df' contains both 'Modified Sentiment' and 'Confidence' columns
print(modified_df)
```

```python
positive_sentiment_df = modified_df[modified_df['Modified Sentiment'] == 'Positive']
print(positive_sentiment_df)
```

```python
sentiment_counts = modified_df['Modified Sentiment'].value_counts()

plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color='skyblue')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.ylim(0,3500)

plt.show()
```

Average Price per Category/Reviews Sum per Category & Bar Chart of Percentiles for Price of Products vs. Reviews Sum Per Product

```python
# Function to convert and handle errors
def convert_to_float(value):
    try:
        # Remove 'S$' and ',' and convert to float
        return float(value.replace('S$', '').replace(',', ''))
    except ValueError:
        return np.nan
    
df['Price2'] = df['Price'].apply(convert_to_float)
df['Price2']
```

```python
# Function to convert and handle errors
def convert_to_float(value):
    try:
        # Remove 'S$' and ',' and convert to float
        return float(value.replace('S$', '').replace(',', ''))
    except ValueError:
        return np.nan
    
category_price_df = df.groupby('Category')['Price'].apply(lambda x: x.apply(convert_to_float))

# Reset the index to get 'Category' as a column instead of the index
category_price_df = category_price_df.reset_index()

# Rename the 'Price' column to 'Price2'
category_price_df.rename(columns={'Price': 'Price2'}, inplace=True)

# Now, 'category_price_df' contains 'Category' and 'Price2'
print(category_price_df)
```

```python
reviews_sum_df = result_df[['Category', 'Reviews Sum']]

# Now, 'reviews_sum_df' contains only 'Category' and 'Reviews Sum'
print(reviews_sum_df)
```

```python
# Calculate the average price per category
average_price_per_category = df.groupby('Category')['Price2'].mean().reset_index()

# Merge the average price information with 'reviews_sum_df' based on 'Category'
reviews_sum_df = pd.merge(reviews_sum_df, average_price_per_category, on='Category')

# Rename the 'Price' column to 'Average Price'
reviews_sum_df.rename(columns={'Price2': 'Average Price'}, inplace=True)

# Now, 'reviews_sum_df' contains 'Category', 'Reviews Sum', and 'Average Price' columns
print(reviews_sum_df)
```

```python
# Create a figure with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Subplot 1: Bar chart for Average Price
axs[0].bar(reviews_sum_df['Category'], reviews_sum_df['Average Price'], color='blue')
axs[0].set_xlabel('Category')
axs[0].set_ylabel('Average Price')
axs[0].set_title('Average Price per Category')

# Subplot 2: Bar chart for Average Rating
axs[1].bar(reviews_sum_df['Category'], reviews_sum_df['Reviews Sum'], color='green')
axs[1].set_xlabel('Category')
axs[1].set_ylabel('Reviews Sum')
axs[1].set_title('Reviews Sum per Category')

# Rotate x-axis labels for better visibility
for ax in axs:
    ax.tick_params(axis='x', labelrotation=0, labelsize=6)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
